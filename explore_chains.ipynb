```python
stopwords_en = stopwords_en = set(stopwords.words('english'))

def stopwords_filter(text):
    filtered = []
    for tok in text.split(' '):
        if tok.lower() not in stopwords_en:
            filtered.append(tok)

    return ' '.join(filtered)
```

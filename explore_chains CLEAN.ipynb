```python
import pickle
import numpy as np
from nltk.corpus import stopwords
```

```python
with open('chains_test_novg_nostopwords.dict', 'rb') as f:
    chains = pickle.load(file=f)
    
with open('chains_test_vg_nostopwords.dict', 'rb') as f:
    chains_d = pickle.load(file=f)
    
with open('annotations/photobook_captions.dict', 'rb') as f:
    captions = pickle.load(f)
```

```python
def print_chains(_image):

    current_game = -1
    current_round = -1

    for c, c_ in zip(chains[_image], chains_d[_image]):

        if current_game != c[0]:
            current_game = c[0]
            print('-'*100)
            
        if current_round != c[1]:
            current_round = c[1]
            print('---')
            print('VG features: {}'.format(c_[5]))
            print('---')

        print('{} {} {}      {:.02f} {}'.format(c[0], c[1], c[2], c[4], c[3]))
        if c[4] != c_[4]:
            print('{} {} {}   VG {:.02f} {}'.format(c_[0], c_[1], c_[2], c_[4], c_[3]))
```

```python
for img in chains:
    img_str = img.split('/')[-1]

    print('{}\n{}'.format(img, '-'*50))
    for capt in captions[img_str]:
        print(capt)
    
    print_chains(img)
```
